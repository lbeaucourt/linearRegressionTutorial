```python
import time
from IPython import display

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
%matplotlib inline
plt.rcParams['figure.figsize'] = [14, 8]
```

```python
x1 = np.random.rand(100)
x2 = (np.random.rand(100)-0.5)

arg1 = np.random.normal(45.2,45.2/10,100)
arg2 = np.random.normal(-23.3,23.3/10,100)

genData = x1*arg1 + x2*arg2

data = pd.DataFrame({'x1':x1,'x2':x2,'y':genData})
data.head()
```

```python
fig,ax = plt.subplots(ncols=2)
ax[0].hist(arg1,bins=20)
ax[1].hist(arg2,bins=20);
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
        
ax.scatter(data.x1,data.x2,data.y,label='J(theta)');
```

```python
fig, ax = plt.subplots(ncols=2)

ax[0].plot(data.x1,data.y,'o')
ax[1].plot(data.x2,data.y,'o');
```

```python
# Dfinir notre hypothse (fonction)

def hypothesis(x,theta):
    return np.dot(x,theta)
```

```python
# On gnre alatoirement une valeur de dpart pour le paramtre theta1 de notre modle

theta = [(np.random.rand()-0.5)*100,(np.random.rand()-0.5)*100]
```

```python
# On dfinit notre fonction de cot: somme quadratique (eg: on somme les carr)

def costFunction(y,yhat):
    return np.power(yhat - y,2).sum()*(2/y.shape[0])
```

```python
yhat = hypothesis(data[['x1','x2']],theta)
costFunction(data.y,yhat)
```

```python
theta
```

```python
thetaRange = np.arange(-100,100,1)
costFctEvol = pd.DataFrame({},index = thetaRange)
for theta0 in thetaRange:
    costFctEvol[theta0] = pd.Series([costFunction(data.y,hypothesis(data[['x1','x2']],[theta0,theta1])) 
                                     for theta1 in thetaRange],index = thetaRange)

costFctEvol.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(costFctEvol.index,costFctEvol.columns,costFctEvol.values,cmap=cm.coolwarm,);
```

```python
def costFctDeriv(x,y,yhat):
    return ((yhat - y)*x.T).sum(axis=1)/y.shape[0]

def gradDescent(x,y,yhat,alpha):
    return -alpha*costFctDeriv(x,y,yhat)
```

```python
gradDescent(data[['x1','x2']],data.y,yhat,0.1)
```

```python
thetaInit = theta
alpha = 1
epsilon = 0.001
```

```python
gradDescentEvol = pd.DataFrame({'theta1':thetaInit[0],
                                'theta2':thetaInit[1],
                                'J':costFunction(data.y,yhat)},index = np.arange(1))
```

```python
def plotCostFunction(fig,data,theta,gradDescentEvol):
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(costFctEvol.index,costFctEvol.columns,costFctEvol.values,cmap=cm.coolwarm,);

    ax.scatter(gradDescentEvol.theta1,gradDescentEvol.theta2,gradDescentEvol.J,'o',color='black')
    ax.legend()

def plotCostFunctionEvol(ax,gradDescentEvol):
    ax.plot(np.arange(gradDescentEvol.shape[0]),gradDescentEvol.J,label='J(theta)')
    ax.legend()
```

```python
# On parametrise deux trois trucs

plt.rcParams['figure.figsize'] = [16, 5]
costFct = 0
count = 0

# Et on se lance dans la boucle: La descente de gradient!
while np.abs(costFunction(data.y,yhat) - costFct)/costFct >= epsilon:
    count += 1
    costFct = costFunction(data.y,yhat)
    theta += gradDescent(data[['x1','x2']],data.y,yhat,alpha)
    yhat = hypothesis(data[['x1','x2']],theta)
    gradDescentEvol = gradDescentEvol.append(pd.DataFrame({'theta1':thetaInit[0],
                                                           'theta2':thetaInit[1],
                                                           'J':costFunction(data.y,yhat)},
                                                          index = np.arange(1)),
                                             ignore_index=True)
    #plotData(ax[0],data,theta,yhat,gradDescentEvol)
    fig, ax = plt.subplots()
    plotCostFunctionEvol(ax,gradDescentEvol)
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(1)
```

```python
print('Convergence: %i steps' % count)
print('theta: [%.2f,%.2f]' % (theta[0],theta[1]))
print('J(theta): %.2f' % costFunction(data.y,hypothesis(data[['x1','x2']],theta)))
```

```python
fig = plt.figure(figsize=[16,10])
plotCostFunction(fig,data,theta,gradDescentEvol)
```
